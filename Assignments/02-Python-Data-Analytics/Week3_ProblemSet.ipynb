b) Write a function that counts how many fruit are mentioned in a given string, narrowing your fruit search to only apple, banana, strawberry, grapes. Think about what kind of function would be most appropriate to use for this and provide the proper amount fo documention. 

```python
def count_fruit_types(inputString: str, fruits_to_search: list) -> dict:
    """Return a dictionary counting how many times each fruit in `inputString` shows up in `fruit_to_search` list.

    Parameters
    ----------
    inputString: str
        string to be searched whether there are fruits in it
    fruits_to_search: list
        list of fruits to search for in inputString

    Returns
    ----------
    dictionary:
        list of fruits {keys} and number of times {value} it appeared in inputString

    """
    return {fruit: inputString.count(fruit) for fruit in fruits_to_search}


def print_dictionary(dict_: dict):
    """A helper function that prints the contents of a dictionary

    Parameters
    ----------
    dict_: dict
        dictionary whose keys and values will be printed out using the print function

    Returns
    ----------
    None
    """

    result = "\nThe following items were found:\n\n"

    for keys, value in dict_.items():
        result += "\t" + str(value) + " " + str(keys) + "\n"
    print(result)


# print out the docs for the 2 functions

print(count_fruit_types.__doc__)
print(print_dictionary.__doc__)

# Option 1:
# print out the result of count_fruit_types dictionary using the print function
print(
    count_fruit_types(
        "There are grapes, apples, oranges, watermelons, grapes, grapes, grapes and bottles of water in the trunk.",
        ["apples", "bananas", "strawberries", "grapes"],
    )
)

# Option 2:
# print out the result of count_fruit_types dictionary using the helper function
print_dictionary(
    count_fruit_types(
        "There are grapes, apples, oranges, watermelons, grapes, grapes, grapes and bottles of water in the trunk.",
        ["apples", "bananas", "strawberries", "grapes"],
    )
)

# a few more tests using different input strings
print_dictionary(
    count_fruit_types("There are koi fish in my backyard.", ["apples", "grapes"])
)  # no matches
print_dictionary(
    count_fruit_types(
        "There are truckloads of bananas for the Zoo gorillas.", ["bananas"]
    )
)  # one match
print_dictionary(
    count_fruit_types("There are grapes in the fridge.", ["grapes", "grapes"])
)  # one match
```

## Problem 2 -  30 pts

a) Write a function that identifies the average weight of penguin that eat fish. Return a float variable. Think about what kind of function would be most appropriate to use for this and provide the proper amount fo documention.  

```python
penguins = pd.read_csv("palmerpenguins_extended.csv")  # create the penguins DataFrame


def get_avg_weight_based_on_diet(
    penguin_dataframe: object,
    column_name: str,
    string_value: str,
    column_to_calculate: str,
) -> float:
    """Return a float indicating the average weight of a penguin based on its diet (NO USER INPUT)

    Uses the built in pandas method DataFrame.query, building the criteria from the arguments in this function

    Parameters
    ----------
    penguin_dataframe: pandas DataFrame
        dataframe with penguin data
    column_name: str
        column name to search for the string string_value (i.e. the "diet" column)
    string_value: str
        string to find in the column_name column
    column_to_calculate: str
        column to calculate the average weight of each penguin based on diet

    Returns
    ----------
    float:
        average weight of each penguin who based on diet

    """
    # build the expression needed for the dataframe.query statement
    expression = column_name + "==" + "'" + string_value + "'"
    result_column = column_to_calculate

    # calculate the average weight and round by 2 decimals, and return as a float
    return round(penguin_dataframe.query(expression)[result_column].mean(), 2)


# print out the docs for the function

print(get_avg_weight_based_on_diet.__doc__)

# this function call is equivalent to running round(penguin_dataframe.query("diet=='fish'")["body_mass_g"].mean(), 2)
avg_weight_fish_eating_penguin = get_avg_weight_based_on_diet(
    penguins, "diet", "fish", "body_mass_g"
)

# print out the value returned by the get_avg_weight_based_on_diet function
print(
    f"The average weight of a penguin on a diet of fish is {avg_weight_fish_eating_penguin} grams."
)
```

b) Rewrite the above function that identifies the average weight of penguin that eat fish, and allow the user to input which type of diety match they want (example fish, krill, etc). Return a tuple with the chosen diety item and the average. Print returning tuple outside of function. Think about what kind of function would be most appropriate to use for this and provide the proper amount fo documention. 

```python
def get_avg_weight_based_on_diet_from_user(
    penguin_dataframe: object, column_name: str, column_to_calculate: str
) -> tuple:
    """Return a tuple indicating the average weight of a penguin based on its diet chosen by user input

    Uses the built in pandas method DataFrame.query, building the criteria from the arguments in this function

    Parameters
    ----------
    penguin_dataframe: pandas DataFrame
        dataframe with penguin data
    column_name: str
        column name to search for the string value (i.e. the "diet" column)
    column_to_calculate: str
        column to calculate the average weight of each penguin based on diet

    Returns
    ----------
    tuple:
        average weight of each penguin who based on diet

    """

    # form the unique diet list based on the DataFrame
    diet_list = [food for food in penguins["diet"].unique()]

    # Keep asking for user valid user input regarding the food

    # initialize loop flag
    invalid_input = True

    while invalid_input:
        try:
            # ask for user input
            print(diet_list)
            response = input("Choose one of the food diets from this list above:")

            if response in diet_list:
                # user has chosen a valid food
                # set flag to end loop
                invalid_input = False

                # build the expression needed for the dataframe.query statement, using response from user
                expression = column_name + "==" + "'" + response + "'"
                result_column = column_to_calculate

                # calculate the average weight and round by 2 decimals, and return as a float
                return (
                    response,
                    round(penguin_dataframe.query(expression)[result_column].mean(), 2),
                )
            else:
                print("Invalid choice, try again.")
        except:
            print("Invalid choice, try again.")


# print out the docs for the function

print(get_avg_weight_based_on_diet_from_user.__doc__)


# call the function, and store the tuple in variable avg_penguin_weight_user_specified
avg_penguin_weight_user_specified = get_avg_weight_based_on_diet_from_user(
    penguins, "diet", "body_mass_g"
)

# print the tuple
print(avg_penguin_weight_user_specified)

# Tested with all four diet food types and it returned the following
# ('fish', 6173.38)
# ('krill', 4627.91)
# ('parental', 3530.87)
# ('squid', 5520.27)
```

## Problem 3 - 50 pts

Write a function that takes in the penguins df, any one of the available numerical columns, species, sex, and weight category and returns a dataframe with the average of that numeric category by sex and weight category. Think about what kind of function would be most appropriate to use for this and provide the proper amount fo documention. 

Example output with made up averages: <br>
Species    Sex      Weight    Average <br>
Chinstrap  Female   Healthy   4685<br>
Chinstrap  Male     Healthy   4985<br>
Chinstrap  Female   Overweight5685<br>
Chinstrap  Male     Overweight5985<br>
Gintoo     Female   Healthy   4675<br>

```python
def get_avg_given_numeric_column(
    penguin_dataframe: object, numeric_column: str
) -> object:
    """Return a DataFrame with columns Species, Sex, Weight, and Average

    Average of the data in the numerical column passed in the function

    Uses the built in pandas method DataFrame.query, building the criteria from the arguments in this function

    Parameters
    ----------
    penguin_dataframe: pandas DataFrame
        dataframe with penguin data
    numeric_column: str
        column to calculate the average

    Returns
    ----------
    DataFrame:
        Columns "Species", "Sex", "Weight", "Average"

    """

    # retrieve the 3 columns from the DataFrame and store in a new DataFrame,
    # sort the new DataFrame and remove the duplicate rows

    subset_dataframe = penguin_dataframe[["species", "sex", "health_metrics"]]
    subset_dataframe = subset_dataframe.sort_values(
        ["species", "sex", "health_metrics"]
    )
    subset_dataframe = subset_dataframe.drop_duplicates()

    # we now have a list of distinct rows to compute averages for

    # Initialize a list of averages
    averages = []

    # go through each row of subset_dataframe to build the criteria for DataFrame.query()
    # then compute the average of numerical_column

    for index, row in subset_dataframe.iterrows():
        # build the expression needed for the dataframe.query statement based on input species, sex, and health_metrics
        expression = "species" + "==" + "'" + row["species"] + "'"
        expression += " & sex" + "==" + "'" + row["sex"] + "'"
        expression += " & health_metrics" + "==" + "'" + row["health_metrics"] + "'"

        # include the numerical column name in the query expression and store the average into a variable
        average = round(penguin_dataframe.query(expression)[numeric_column].mean(), 2)

        # add the calculated average to a list that will be used for a new column in subset_dataframe
        averages.append(average)

    # create a new colunm Average in subset_dataframe and insert the list of average values in each row

    subset_dataframe["Average"] = averages

    # rename column names to match the instructor's example
    subset_dataframe.rename(
        columns={"species": "Species", "sex": "Sex", "health_metrics": "Weight"},
        inplace=True,
    )

    # return dataframe
    return subset_dataframe


# print out the docs for the function

print(get_avg_given_numeric_column.__doc__)

# test the function

numerical_column_average = get_avg_given_numeric_column(penguins, "body_mass_g")
print("Calculate body_mass_g averages")
print()
print(numerical_column_average)  # dataframe using body mass
print()

print("Calculate bill_length_mm averages")
print()
numerical_column_average = get_avg_given_numeric_column(penguins, "bill_length_mm")
print(numerical_column_average)  # dataframe using bill length
print()

print("Calculate bill_depth_mm averages")
print()
numerical_column_average = get_avg_given_numeric_column(penguins, "bill_depth_mm")
print(numerical_column_average)  # dataframe using bill depth
print()

print("Calculate flipper_length_mm averages")
print()
numerical_column_average = get_avg_given_numeric_column(penguins, "flipper_length_mm")
print(numerical_column_average)  # dataframe using flipper length
print()
```
