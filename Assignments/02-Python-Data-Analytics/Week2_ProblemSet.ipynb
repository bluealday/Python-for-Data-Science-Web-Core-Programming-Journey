## Problem 1 - 10 pts

How many of each penguin species are in the dataset?

```python
penguins = pd.read_csv('palmerpenguins_extended.csv') # create the penguins DataFrame

# use a list comprehension to count the number of penguins that belong to a unique list of species using a dictionary
# and the DataFrame.value_counts function
penguins_count = [{penguin : penguins['species'].value_counts()[penguin]} for penguin in penguins['species'].unique()]

# print the penquin count for each species from the list of dictionaries
print("Option 1: Show dictionary")
print(f"The number of penguins for each species is: {penguins_count}")

# or if they need to be in separate lines?
print()
print("Option 2: Print individual values")


for dict_ in penguins_count:
    for keys, value in dict_.items():
        print(f'There are {value} {keys} penguins.')
#penguins.describe()
```

## Problem 2 -  50 pts

a) What is the min, max, and average of each of the numeric columns within the dataset?

```python
def get_dictionary_Data(l, typedata):
    # function that takes a list of dictionaries l and string typedata and prints out the column name and value 
    print()
    for dict_ in l:
        for keys, value in dict_.items():
            print(f'The {typedata} value for column {keys} is {value}.')


# create a list of numeric columns
numeric_columns = ['bill_length_mm' ,'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'year']

# create a list of dictionaries of min, max, and mean (avg) values for each numeric column using built in functions
# for readability, round the average to 2 decimal places

# also note that techically the year column is numeric and its average is not a whole number. For simplicity, 
# the year average was also set to 2 decimal places

min_values = [{column: penguins[column].min()} for column in numeric_columns]
max_values = [{column: penguins[column].max()} for column in numeric_columns]
mean_values = [{column: round(penguins[column].mean(), 2)} for column in numeric_columns]

# print the min, max, and mean (avg) values for each column from the list of dictionaries
print("Option 1: Show dictionary")
print(f"Min: {min_values}")
print()
print(f"Max: {max_values}")
print()
print(f"Avg: {mean_values}")
print()

# or if they need to be in separate lines?
print("Option 2: Print individual values")
get_dictionary_Data(min_values, 'minimum')
get_dictionary_Data(max_values, 'maximum')
get_dictionary_Data(mean_values, 'average')

        
# In case there will be more numeric type columns in the future, its just one line of code
# to add new them columns to the list numeric_columns
```

b) Convert the numeric (mm) columns into centimeters.

```python
# create a list of numeric columns that have mm
numeric_columns_mm = ['bill_length_mm' ,'bill_depth_mm', 'flipper_length_mm']

# multiply the end values by 0.1 to convert mm to cm
# change the corresponding column labels from mm to cm
# for readability, round the average to 2 decimal places

min_values_cm = [{column.replace('_mm', "_cm"): penguins[column].min() * 0.1} for column in numeric_columns_mm]
max_values_cm = [{column.replace('_mm', "_cm"): penguins[column].max() * 0.1} for column in numeric_columns_mm]
mean_values_cm = [{column.replace('_mm', "_cm"): round(penguins[column].mean() * 0.1, 2)} for column in numeric_columns_mm]

# print the min, max, and mean (avg) values for each mm column that was converted to cm
print("Option 1: Show dictionary")
print(f"Min (cm): {min_values_cm}")
print()
print(f"Max (cm): {max_values_cm}")
print()
print(f"Avg (cm): {mean_values_cm}")
print()

print("Option 2: Print individual values")
get_dictionary_Data(min_values_cm, 'minimum')
get_dictionary_Data(max_values_cm, 'maximum')
get_dictionary_Data(mean_values_cm, 'average')

# In case there will be more cm calculations in the future, its just one line of code
# to add new mm columns to the list
```

c) Convert body mass to pounds.

```python
# create a list of numeric columns that have grams
numeric_columns_g = ['body_mass_g']

# multiply the end values by 0.00220462 to convert g to lbs
# change the corresponding column labels from g to lbs
# will not round any values since data fits well in the window below

min_values_lbs = [{column.replace('_g', "_lbs"): penguins[column].min() * 0.00220462} for column in numeric_columns_g]
max_values_lbs = [{column.replace('_g', "_lbs"): penguins[column].max() * 0.00220462} for column in numeric_columns_g]
mean_values_lbs = [{column.replace('_g', "_lbs"): penguins[column].mean() * 0.00220462} for column in numeric_columns_g]

# print the min, max, and mean (avg) values for each gram column that was converted to lbs
print("Option 1: Show dictionary")
print(f"Min (lbs): {min_values_lbs}")
print()
print(f"Max (lbs): {max_values_lbs}")
print()
print(f"Avg (lbs): {mean_values_lbs}")
print()

print("Option 2: Print individual values")
get_dictionary_Data(min_values_lbs, 'minimum')
get_dictionary_Data(max_values_lbs, 'maximum')
get_dictionary_Data(mean_values_lbs, 'average')

# In case there will be more lbs calculations in the future, its just one line of code
# to add new gram columns to the list
```

## Problem 3 - 20 pts

Create a new column that calculates the difference in body mass (g) of each penguin and the average weight of all penguins. 

```python
# use dict mean_values and key "body_mass_g" to subtract from each penguin's mass

# extract mean value of body_mass_g and exit from loop once it is found
for dict_ in mean_values:
    if ("body_mass_g" in dict_.keys()):
        avg_weight = dict_.get("body_mass_g")
        break
        
# create new column body_mass_diff_g as instructed and write the changes back into our Dataframe penguins
penguins = penguins.assign(body_mass_diff_g = penguins.body_mass_g - avg_weight)

# show DataFrame on screen
penguins
```

## Problem 4 - 20 pts

Create a new column classifies each penguin into quartiles by body mass (g). Example, if the penguins bodyw eight falls within the first 25

```python
# create a list of conditions to classify body mass into 25, 50, 75, and 100 percentiles

# recall 
# min was 2477, max was 10549

# extract min and max value of body_mass_g and exit from loop once it is found
for dict_ in min_values:
    if ("body_mass_g" in dict_.keys()):
        min_weight = dict_.get("body_mass_g")
        break

for dict_ in max_values:
    if ("body_mass_g" in dict_.keys()):
        max_weight = dict_.get("body_mass_g")
        break

# divide the max/min range into 4
quarter = (max_weight - min_weight) * 0.25

#define the conditions and values based on the lecture
conditions = [
    (penguins['body_mass_g'] <= (2 * quarter) -1),  # 25%
    (penguins['body_mass_g'] >= (2 * quarter)) & (penguins['body_mass_g'] <= (3 * quarter) -1),  # 50%
    (penguins['body_mass_g'] >= (3 * quarter)) & (penguins['body_mass_g'] <= (4 * quarter) -1),  # 75%
    (penguins['body_mass_g'] >= (4 * quarter)) # 100%
]

values = ['0-25%', '26-50%', '51-75%', '76-100%']

# create a new column for the percentiles of body mass
penguins['mass_categories'] = np.select(conditions, values)

# display updated DataFrame, using tail in this case displays all 4 percentile ranges by coincidence 
penguins.tail(50)


```
