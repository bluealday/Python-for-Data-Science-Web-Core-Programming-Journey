# Week 7 Discussion Topic: Koi in the Pond

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

fig, ax = plt.subplots()

koi_roster = '''
Black Diamond
Black Diamond
Doitsu Kawarimono
Doitsu Kohaku
Doitsu Sanke
Doitsu Showa
Ginrin Asagi
Karashigoi
Ki Matsuba
Kikokuryu
Ki Kikokuryu
Kin Matsuba 
Kohaku
Kujaku
Kujaku
Showa
Maruten Kohaku
Maruten Kohaku
Maruten Sanke
Showa
Shiro Utsuri
Sanke
Showa
Showa
Shusui
Kohaku
Kohaku
Kin Ki Utsuri
Mukashi Ogon
Sanke
Doitsu Kujaku
Shiro Utsuri
Shusui
Shusui
Sanke
Tancho Kohaku
Goshiki
Gin Matsuba 
Kogane Ochiba
Yamabuki Ogon
Platinum Ogon
Kujaku
Showa
Ginrin Chagoi
Golden Corn
Doitsu Bekko
Kin Shiro Utsuri
Yamabuki Ogon
Sanke with Kuchibeni
Golden Corn
Golden Corn
Kirin
'''

big_one = "Kohaku"
big_two = "Sanke"
big_three = "Showa"

gosanke_combined = []
individual_gosanke = []
non_gosanke = []


#################################################################################################
# Chart 1: Get koi varieties, clean data, use a sorted set for chart "Total Koi Fish in the Pond"

varieties = koi_roster.splitlines()

sorted(varieties)
if varieties[0] == '':
    varieties.remove('')

varieties = sorted(set(varieties))

# Get koi counts per variety
counts = ([koi_roster.count(koi) for koi in varieties])
bar_colors = list(mcolors.CSS4_COLORS)

ax.barh(varieties, counts, label=varieties, color=bar_colors)
ax.set_title('Total Koi Fish in the Pond')
ax.set_xlabel('Count per Variety')
ax.set_ylabel('Variety')
figure = plt.gcf() # get current figure
figure.set_size_inches(12, 7.68)
plt.savefig('Week 7 - matplotlib - Pond Koi Distribution.png', dpi = 100)
plt.show()

#################################################################################################


#################################################################################################
# Chart 2: "Grouped by Gosanke variety in the Pond"

fig, ax = plt.subplots() 


# If current koi is one of the big three, build the gosanke_combined list

group_gosanke = koi_roster.splitlines()

for koi in group_gosanke:
    if koi.find("Kohaku") != -1: 
        gosanke_combined.append(big_one)
    elif koi.find("Sanke") != -1:
        gosanke_combined.append(big_two)
    elif koi.find("Showa") != -1:
        gosanke_combined.append(big_three)

if gosanke_combined[0] == '':
    gosanke_combined.remove('')

gosanke_combined.sort()

# Get koi counts per gosanke
gosanke_combined_counts = ([gosanke_combined.count(koi) for koi in gosanke_combined])
bar_colors = list(mcolors.CSS4_COLORS)

ax.bar(gosanke_combined, gosanke_combined_counts, color=bar_colors)
ax.set_title('Grouped by Gosanke variety in the Pond')
ax.set_ylabel('Count by Variety')
figure = plt.gcf() # get current figure
figure.set_size_inches(8, 6)
plt.ylim(1, 7)
plt.savefig('Week 7 - matplotlib - Top Three Koi Show Varieties.png', dpi = 100)
plt.show()

#################################################################################################


#################################################################################################
# Chart 3: "Non Gosanke Koi Fish in the Pond"
fig, ax = plt.subplots()

for koi in varieties:
    if not (koi.find("Kohaku") != -1 or koi.find("Sanke") != -1 or koi.find("Showa") != -1):
        non_gosanke.append(koi)

# Get koi counts per non_gosanke
non_gosanke_counts = ([koi_roster.count(koi) for koi in non_gosanke])
bar_colors = list(mcolors.CSS4_COLORS)

ax.barh(non_gosanke, non_gosanke_counts, label=non_gosanke, color=bar_colors)
ax.set_title('Non Gosanke Koi Fish in the Pond')
ax.set_xlabel('# Koi')
ax.set_ylabel('Varieties')
figure = plt.gcf() # get current figure
figure.set_size_inches(16, 12)
plt.savefig('Week 7 - matplotlib - Non Gosanke Varieties.png', dpi = 100)
plt.show()

#################################################################################################


#################################################################################################
# Chart 4: "Individual Gosanke Koi Fish in the Pond"

fig, ax = plt.subplots()

find_gosanke = koi_roster.splitlines()

for koi in find_gosanke:
    if (koi.find("Kohaku") != -1 or koi.find("Sanke") != -1 or koi.find("Showa") != -1):
        individual_gosanke.append(koi)

if individual_gosanke[0] == '':
    individual_gosanke.remove('')

individual_gosanke.sort()

# Get koi counts per indvidual gosanke type
individual_gosanke_counts = ([individual_gosanke.count(koi) for koi in individual_gosanke])

bar_colors = list(mcolors.CSS4_COLORS)

plot = plt.figure
ax.barh(individual_gosanke, individual_gosanke_counts, label=individual_gosanke, color=bar_colors)
ax.set_title('Individual Gosanke Koi Fish in the Pond')
ax.set_xlabel('# Koi')
figure = plt.gcf() # get current figure
figure.set_size_inches(16, 12)
plt.savefig('Week 7 - matplotlib - Top Three Koi Show Varieties 2.png', dpi = 100)
plt.show()
```
